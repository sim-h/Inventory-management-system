```python
#medicine dataset
import pandas as pd
import numpy as np
md=pd.read_csv('medicine.csv')
md
```

```python
#generating demand for 365 days
ls=list(md['Medicine'])
dat={}
k=list(zip(md['Mean Demand'],md['Std Dev of dem']))
for l in range(len(ls)):
    dat[ls[l]]=np.random.normal(k[l][0],k[l][1],365)
f=pd.DataFrame(dat)
f
```

```python
#generating annual demand from the demand data of 365 days
f.loc['Annual Demand']=f.sum()
g=pd.DataFrame(f.loc['Annual Demand'],index=f.columns,columns=['Annual Demand'])
g=g.reset_index()

g=g.rename(columns={'index':'Medicine'})
md=pd.merge(md,g,on='Medicine')
md
```

```python
#complete dataframe with demand of 365 days
md
```

```python
#allocation of medicines to test centers
sup=pd.read_csv('E:\center.csv')
sup
```

```python
#allocation of medicines to suppliers
cen=pd.read_csv('E:\Supplier.csv')
cen
```

```python
#mapping the medicines to the test centers
md=pd.merge(md,sup,on='Medicine')
md
```

```python
#mapping the suppliers
md=pd.merge(md,cen,on='Medicine')
md
```

```python
#calculating the total lead time
md['Total lt']=md['Lead time']+md['LT']
md
```

```python
#calculating the standard deviation of lead time
md['Total sdev']=(((md['Sdev'])**2)+((md['std dev'])**2))**0.5
md
```

```python
#calculation of rop using lead time uncertainity
from scipy.stats import norm
md['ROP']=((norm.ppf(0.9,loc=md['Mean Demand'],scale=md['Std Dev of dem']))*(((md['Total lt']*((md['Std Dev of dem'])**2))+(((md['Mean Demand'])**2)*((md['Total sdev'])**2)))**0.5))+md['Mean Demand']*md['Total lt']
md
```

```python
#calculation of EOQ using the model-lots ordered and delivered jointly using selected subsets
md['ni']=((md['holding cost']*md['Price']*md['Annual Demand'])+(2*(4000+md['Or.Cost'])))**0.5
md['nii']=((md['holding cost']*md['Price']*md['Annual Demand'])+(2*(md['Or.Cost'])))**0.5
md['mi']=md['ni']//md['nii']
md['num']=md['holding cost']*md['Price']*md['Annual Demand']*md['mi']
md['deno']=md['Or.Cost']/md['mi']
b=md['num'].apply('sum')
c=md['deno'].apply('sum')
c=(c+4000)*2
n=(b/c)**0.5
md['Or.Freq']=n//md['mi']
md
```

```python
#final datset with all parameters
nd=md.drop(['ni','nii','mi','num','deno'],axis=1)
nd
```

```python
#creating an EOQ column in the first(md) dataset
nd['EOQ']=nd['Annual Demand']//nd['Or.Freq']
md['EOQ']=nd['EOQ'].copy()
md
```

```python
#setting the index as medicine
nd=nd.set_index('Medicine')
nd

        
```

```python
#generating the stock left dataset
jk=pd.DataFrame(columns=nd.index.values,index=range(365))
j=[]
for t in nd.index.values:
    s=list(f[t])
    k=[]
    x=nd.loc[t]['EOQ']
    l=0
    for i in range(len(s)-1):
        x=x-s[i]
        k.append(x)
        if(x==0):
            x=nd.loc[t]['EOQ'].values
            l=l+1
            
        elif (i!=len(s)-1 and x-s[i+1] < 0):
            x=nd.loc[t]['EOQ']
            l=l+1
            
       
    j.append(l)
    jk[t]=k
nd['act_Orfreq']=j
j
```
